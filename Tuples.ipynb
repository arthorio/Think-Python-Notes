```python
def min_max(t):
    return min(t), max(t)
```

```python
def mean(*args):
    return sum(args) / len(args)

mean(1, 2, 6, 4, 5, 3)
```

```python
def trimmed_mean(*args):
    low, high = min_max(args)
    trimmed = list(args)
    trimmed.remove(low)
    trimmed.remove(high)
    return mean(*trimmed)
```

```python
def value_counts(string):
    counter = {}
    for letter in string:
            counter[letter] = counter.get(letter, 0) + 1
    return counter
```

```python
counter = value_counts('banana')
counter
```

```python
items = counter.items()
items
```

```python
sorted(items)
```

```python
def second_element(t):
    return t[1]
```

```python
sorted_items = max(items, key=second_element)
sorted_items
```

```python
def invert_dict(d):
    new = {}
    for key, value in d.items():
        if value not in new:
            new[value] = [key]
        else:
            new[value].append(key)
    return new

d =  value_counts('parrot')
invert_dict(d)
```

```python
letters = 'abcdefghijklmnopqrstuvwxyz'
numbers = range(len(letters))
letter_map = dict(zip(letters, numbers))
```

```python
letters[1]
```

```python
def shift_word(word, number):
    new_word = []
    for letter in word:
        index = (letter_map[letter] + number)  % len(letters)
        new_word.append(letters[index])
    return ''.join(new_word)
```

```python
shift_word('cheer', 7)
```

```python
shift_word('melon', 16)
```

```python
def most_frequent_letters(word):
    word_dict = {}
    for letter in word:
        word_dict[letter] = word_dict.get(letter, 0) + 1
    return list(sorted(word_dict.items(), key=lambda x: x[1], reverse=True))
```

```python
word_list = []

for line in open('words.txt'):
    word = line.strip()
    word_list.append(word)

def anagram_list(lista):
    word_dict = {}
    for word in lista:
        sorted_word = ''.join(sorted(word))
        if sorted_word not in word_dict:
            word_dict[sorted_word] = []

        word_dict[sorted_word].append(word)

    return word_dict
```

```python
def word_distance(word1, word2):
    differ = 0
    for l1, l2 in zip(word1, word2):
        if l1 != l2:
            differ += 1
    return differ
```

```python
print(word_distance("apple", "apric"))   # 3
print(word_distance("house", "mouse"))   # 1
print(word_distance("light", "night"))   # 1
print(word_distance("train", "brain"))   # 1
print(word_distance("spark", "spork"))   # 1
```

```python
from itertools import combinations

def metathesis_pair(word_list):
    word_dict = {}
    word_dict = anagram_list(word_list)
    
    for anagram_group in word_dict.values():  # only the lists of anagrams
        # compare each pair of words in the list
        if len(anagram_group) > 1:
            for word1, word2 in combinations(anagram_group, 2):
                if word_distance(word1, word2) == 2:
                    print(word1, word2)
```

```python
metathesis_pair(word_list)
```
